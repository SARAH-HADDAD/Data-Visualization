```python
import matplotlib.pyplot as plt 
import pandas as pd

df=pd.read_csv("us-cities.csv")

plt.figure(figsize=(10,10))
plt.barh(df['place'],df['population'],color="skyblue")
plt.xlabel("population")
plt.ylabel("State")
plt.title("Population by  State")
plt.grid(True)
plt.xticks(rotation=90)
plt.show
```

```python
plt.figure(figsize=(15,15))
plt.scatter(df['lon'],df["lat"],c=df['population'],s=[pop/500 for pop in df["population"]])

for i, place in enumerate(df["place"]):
    plt.text(df["lon"][i],df["lat"][i],place, fontsize=6,ha="center",va="bottom")
plt.colorbar(label="Population")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("City location and population")
plt.show()
```

```python
data=pd.read_csv('us-productivity.csv')

plt.hist(data["value"],bins=3,edgecolor='k',color="skyblue")
plt.xlabel("Procuctivity Values")
plt.ylabel("Frequency")
plt.title("Histogram of Productivity Values")
plt.grid(True)
plt.show()
```
